```python
import pandas as pd
```

```python
df_obama = pd.read_csv("obama.csv")
df_obama
```

```python
df_trump = pd.read_csv("realdonaldtrump.csv")
df_trump
```

```python
#adding year / month column to obama dataset to account for the fact that the Trump dataset only goes up to 2020
df_obama["year"] = pd.DatetimeIndex(df_obama["Timestamp"]).year
df_obama["month"] = pd.DatetimeIndex(df_obama["Timestamp"]).month
df_obama
```

```python
df_obama = df_obama[(df_obama["year"] <= 2020) & (df_obama["month"] <= 6)]
df_obama
#All set to start analysis.
```

```python
obama_wife = df_obama["Embedded_text"].str.contains("Michelle").sum()
obama_wife
```

```python
trump_wife = df_trump["content"].str.contains("Melania").sum()
trump_wife
```
