```python
#extracting date information from date columns
df["Amount received year"] = pd.DatetimeIndex(data = df["Amount Received Date"]).year
df["Amount received month"] = pd.DatetimeIndex(data = df["Amount Received Date"]).month
df["Amount received day"] = pd.DatetimeIndex(data = df["Amount Received Date"]).day
df["Report received year"] = pd.DatetimeIndex(data = df["Report Received Date"]).year
df["Report received month"] = pd.DatetimeIndex(data = df["Report Received Date"]).month
df["Report received day"] = pd.DatetimeIndex(data = df["Report Received Date"]).day
df
```

```python
#Making dollar amount numeric data
df["Amount"] = df["Amount"].replace(to_replace = "\$", value = "", regex = True)
df["Amount"] = df["Amount"].replace(to_replace = ",", value = "", regex = True)
df["Amount"] = df["Amount"].astype(float)
df
```

```python
#calculating the delay between receiving the money and reporting the money

data = []
for i in range (len(df)):
    difference = df["Report received month"][i] - df["Amount received month"][i]
    if difference < 0: 
        delay = 12 + difference
    else: 
        delay = difference
    d = {"Report delay months" : delay}
    data.append(d)
df["Report delay months"] = data
```
