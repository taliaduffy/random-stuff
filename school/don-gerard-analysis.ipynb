```python
import pandas as pd
df = pd.read_csv("don-gerard.csv")
df
```

```python
#extracting date information from date columns
df["RcvdYear"] = pd.DatetimeIndex(data = df["RcvdDate"]).year
df["RcvdMonth"] = pd.DatetimeIndex(data = df["RcvdDate"]).month
df["RcvdDay"] = pd.DatetimeIndex(data = df["RcvdDate"]).day
df
```

```python
df_year = df.groupby("RcvdYear").agg("sum").reset_index()
df_year
```
