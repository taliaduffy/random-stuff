```python
import pandas as pd
```

```python
df = pd.read_csv("taylorswift.csv")
df
```

```python
df_album = df.groupby("Album").agg("mean").reset_index()
df_album
```

```python
df_album["Year"] = [2014, 2008, 2019, 2012, 2010, 2006, 2020, 2020, 2017]
df_album
```

```python
df_album_year = df_album_ranked = df_album.sort_values("Year", ascending = True).reset_index()
df_album_year
```

```python
df_album_year.plot.bar(x = "Album", y = "Score")
```

```python
df_album_ranked = df_album.sort_values("Score", ascending = False).reset_index()
df_album_ranked
```
